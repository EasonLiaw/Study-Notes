## Skewness (3rd Business Moment)

Skewness represents measure of asymmetry in distribution.



Compared to normal distribution, skewness of normal distribution is usually close to zero.

Skewness of variable can also be computed in Python as follows:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as st
import matplotlib.pyplot as plt
data = pd.read_csv(r'C:\Users\yixia\OneDrive\Desktop\Data Science Notes\Python\Data_sources\Indian_cities.csv')
```

```python
# Heavily right skewed distribution
sns.displot(data, x='population_total',kind='hist',bins=10)
plt.title('Distribution of population_total')
plt.show()
```

```python
# Left skewed distribution
sns.displot(data, x='effective_literacy_rate_male',kind='hist')
plt.title('Distribution of effective_literacy_rate_male')
plt.show()
```

```python
# Symmetric distribution
sns.displot(data, x='sex_ratio',kind='hist')
plt.title('Distribution of sex_ratio')
plt.show()
```

```python
print("Skewness of population_total:",data.drop_duplicates()['population_total'].skew())
print("Skewness of effective_literacy_rate_male:",data.drop_duplicates()['effective_literacy_rate_male'].skew())
print("Skewness of sex_ratio:",data.drop_duplicates()['sex_ratio'].skew())
```

## Kurtosis (4th Business Moment)

Kurtosis represents the degree of outlier presence and measure of distribution peakness.



Compared to normal distribution, kurtosis of normal distribution is usually close to 3.

If using Fisher's definition, kurtosis of normal distribution is close to 0 (Python's kurtosis function uses Fisher's definition)

Kurtosis of variables can also be computed in Python as follows:

```python
print("Kurtosis of population_total:",data.drop_duplicates()['population_total'].kurt())
print("Kurtosis of effective_literacy_rate_male:",data.drop_duplicates()['effective_literacy_rate_male'].kurt())
print("Kurtosis of sex_ratio:",data.drop_duplicates()['sex_ratio'].kurt())
```

## 5 Number Summary (Box plots)

5 number summary is a specific set of descriptive statistics that provides the following summary:

1. Minimum
2. 25th percentile (1st quartile - Q1)
3. 50th percentile
4. 75th percentile (3rd quartile - Q3)
5. Maximum

xth percentile represents the threshold at which x% of data is below the threshold.

<b>Percentile = (Number of values below threshold/Total count) * 100</b>

<b>Threshold percentile = (Percentile/100) * (Total count + 1)</b>

The 5 number summary can be represented in a boxplot to identify data symmetry (skewness) and outliers if exist.

Outliers of a given distribution are determined based on values that do not fall within the range of lower threshold and upper threshold as follows for different distribution types:



Note that outliers shown in boxplot are determined based on interquartile range. For normal distributions, values that are 3 standard deviations away from the mean should be considered as outliers.

```python
normal_values = np.random.normal(size=1000)
```

```python
# 5 number summary
pd.Series(normal_values).describe()
```

```python
sns.boxplot(x = normal_values)
plt.show()
```

```python
# Outliers from Gaussian distribution
lower_threshold = np.mean(normal_values) - 3 * np.std(normal_values)
upper_threshold = np.mean(normal_values) + 3 * np.std(normal_values)
outliers = normal_values[(normal_values > upper_threshold) | (normal_values < lower_threshold)]
outliers
```

```python
non_normal_values = np.random.pareto(2,size=1000)
```

```python
# 5 number summary
pd.Series(non_normal_values).describe()
```

```python
sns.boxplot(x = non_normal_values)
plt.show()
```

```python
# Outliers from Non-Gaussian distribution
Q1 = np.percentile(non_normal_values,25)
Q3 = np.percentile(non_normal_values,75)
IQR = Q3 - Q1
lower_threshold = Q1 - 1.5 * IQR
upper_threshold = Q3 + 1.5 * IQR
outliers = non_normal_values[(non_normal_values > upper_threshold) | (non_normal_values < lower_threshold)]
outliers
```

## Hypothesis Testing

Hypothesis testing is used to identify whether random chance might be responsible for an observed effect.

Hypothesis testing usually consists of the following steps:

1. Defining null and alternative hypothesis about a population parameter
2. Collecting sample data
3. Compute test statistic/p-value
4. Decision making of whether to reject null hypothesis or not

<b>Null hypothesis</b> refers to hypothesis that effects observed is due to random chance.

<b>Alternative hypothesis</b> refers to hypothesis that effects observed is not due to random chance.

Hypothesis testing usually involves risk of incorrect decision making of whether to reject null hypothesis, which its summary results can be represented in a confusion matrix form (also relevant for classification problems)



<b> Note that significance level (alpha) is usually defined by domain expertise. </b>

<b>For one continuous variable and one discrete factor<b>:

1. Comparing 1 discrete variable with external standard population:



2. Comparing 2 discrete variables with each other



3. Comparing more than 2 discrete variables with each other


    
Note that the tests above requires normality test and variance test to determine type of test to use for hypothesis testing:
    
<b>Normality test: Anderson test</b>

<b>Variance test: Levene test</b>

<b>For one discrete variable and one discrete factor</b>:



```python
data = pd.read_csv('datasets/Fraud_check.csv')
sub_data = pd.get_dummies(data)
sub_data.info()
```

```python
# Normality test
st.anderson(sub_data["Work.Experience"],'norm')
```

From Anderson test, since test statistic > critical values at 5% significance level, there is sufficient evidence to conclude that the work experience variable is not normally distributed.

```python
# Variance test
st.levene(sub_data["Work.Experience"], sub_data['Urban_YES'])
```

From levene test, since p-value is less than 0.05, there is sufficient evidence to conclude that there is a difference in variance between urban samples and non-urban samples.

The following examples below illustrate the use of various statistical tests as highlighted in previous diagrams above:

#### 1. One Sample Z test (Normal distribution with known Population Standard Deviation)

Null hypothesis: Average work experience of population is 15.

Alternative hypothesis: Average work experience of population is not 15.

```python
from statsmodels.stats.weightstats import ztest
# One sample z test
result = ztest(sub_data["Work.Experience"],value=15)
print('Test statistic is',result[0])
print('P value is',result[1])
```

#### 2. One Sample T test (Normal distribution with unknown Population Standard Deviation)

Null hypothesis: Average work experience of population is 15.

Alternative hypothesis: Average work experience of population is not 15.

```python
# One sample t test
result = st.ttest_1samp(sub_data["Work.Experience"],15)
print('Test statistic is',result[0])
print('P value is',result[1])
```

#### 3. One sample sign test (Non-normal distribution)

Null hypothesis: Average work experience of population is 15.

Alternative hypothesis: Average work experience of population is not 15.

```python
# One sample sign test
result = st.wilcoxon(sub_data["Work.Experience"],np.repeat(15,len(sub_data)))
print('Test statistic is',result[0])
print('P value is',result[1])
```

#### 4. Sample T test for Equal Variance (Normal distribution with independent external conditions)

Null hypothesis: There is no difference in average work experience between urbans and non-urbans.

Alternative hypothesis: There is difference in average work experience between urbans and non-urbans.

```python
urbans = sub_data[sub_data["Urban_YES"]==1]
non_urbans = sub_data[sub_data["Urban_YES"]==0]
```

```python
# 2 sample T test for equal variance
result = st.ttest_ind(urbans['Work.Experience'],non_urbans['Work.Experience'])
print('Test statistic is',result[0])
print('P value is',result[1])
```

#### 5. Sample T test for Unequal Variance (Normal distribution with independent external conditions)

Null hypothesis: There is no difference in average work experience between urbans and non-urbans.

Alternative hypothesis: There is difference in average work experience between urbans and non-urbans.

```python
# 2 sample T test for unequal variance
result = st.ttest_ind(urbans['Work.Experience'],non_urbans['Work.Experience'],equal_var=False)
print('Test statistic is',result[0])
print('P value is',result[1])
```

#### 6. Paired T Test (Normal distribution with dependent external conditions)

Null hypothesis: There is no difference in average work experience between urbans and undergraduates.

Alternative hypothesis: There is difference in average work experience between urbans and undergraduates.

```python
# Paired T Test
result = st.ttest_rel(sub_data['Undergrad_YES'],sub_data['Urban_YES'])
print('Test statistic is',result[0])
print('P value is',result[1])
```

#### 7. Mann Whitney Test (Non-normal distribution)

Null hypothesis: There is no difference in average work experience between urbans and non-urbans.

Alternative hypothesis: There is difference in average work experience between urbans and non-urbans.

```python
# Mann Whitney Test
result = st.mannwhitneyu(urbans['Work.Experience'],non_urbans['Work.Experience'])
print('Test statistic is',result[0])
print('P value is',result[1])
```

#### 8. One Way ANOVA Test (Normal distribution with equal variance)

Null hypothesis: There is no difference in average work experience between single, married and divorced individuals.

Alternative hypothesis: There is no difference in average work experience between single, married and divorced individuals.

```python
# One way ANOVA Test
result = st.f_oneway(sub_data[sub_data['Marital.Status_Married']==1]['Work.Experience'],sub_data[sub_data['Marital.Status_Single']==1]['Work.Experience'],sub_data[sub_data['Marital.Status_Divorced']==1]['Work.Experience'])
print('Test statistic is',result[0])
print('P value is',result[1])
```

#### 9. ANOM Test (Normal distribution with unequal variance)

While there is ANOM function in R, Python uses "Alexander Govern" test as an alternative to ANOM test.

Null hypothesis: There is no difference in average work experience between single, married and divorced individuals.

Alternative hypothesis: There is no difference in average work experience between single, married and divorced individuals.

```python
# Alexander Govern Test
result = st.alexandergovern(sub_data[sub_data['Marital.Status_Married']==1]['Work.Experience'],sub_data[sub_data['Marital.Status_Single']==1]['Work.Experience'],sub_data[sub_data['Marital.Status_Divorced']==1]['Work.Experience'])
print('Test statistic is',result.statistic)
print('P value is',result.pvalue)
```

Note that ANOVA and Alexander Govern test does not compare pairwise difference in sample statistic between multiple groups to reduce risk of interesting observations due to random chance.

#### 10. Mood's Median Test (Non-normal distribution)

Null hypothesis: There is no difference in average work experience between single, married and divorced individuals.

Alternative hypothesis: There is no difference in average work experience between single, married and divorced individuals.

```python
# Testing for difference in work experience between married and single individuals
result = st.mood(sub_data[sub_data['Marital.Status_Married']==1]['Work.Experience'],sub_data[sub_data['Marital.Status_Single']==1]['Work.Experience'])
print('Test statistic is',result[0])
print('P value is',result[1])
```

```python
# Testing for difference in work experience between married and divorced individuals
result = st.mood(sub_data[sub_data['Marital.Status_Married']==1]['Work.Experience'],sub_data[sub_data['Marital.Status_Divorced']==1]['Work.Experience'])
print('Test statistic is',result[0])
print('P value is',result[1])
```

```python
# Testing for difference in work experience between single and divorced individuals
result = st.mood(sub_data[sub_data['Marital.Status_Divorced']==1]['Work.Experience'],sub_data[sub_data['Marital.Status_Single']==1]['Work.Experience'])
print('Test statistic is',result[0])
print('P value is',result[1])
```

#### 11. One proportion test (Comparing proportions between 1 population and external standard)

Null hypothesis: Proportion of urbans in the population is 0.5

Alternative hypothesis: Proportion of urbans in the population is not equal to 0.5

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
# One proportion test
result = proportions_ztest(sub_data['Urban_YES'].value_counts()[1],len(sub_data),0.5)
print('Test statistic is',result[0])
print('P value is',result[1])
```

#### 12. Two Proportion Test (Comparing proportions between 2 populations)

Null hypothesis: Proportion of urbans in the population is the same between undergraduates and non-undergraduates.

Alternative hypothesis: Proportion of urbans in the population is different between undergraduates and non-undergraduates.

```python
# Cross tabulation
cross_data = pd.crosstab(index=sub_data['Undergrad_YES'],columns = sub_data['Urban_YES'])
cross_data
```

```python
# Two proportion test
result = proportions_ztest([cross_data[0][1],cross_data[1][1]],[np.sum(cross_data[0]),np.sum(cross_data[1])])
print('Test statistic is',result[0])
print('P value is',result[1])
```

## Statistical Significance and P-Values

Statistical significance measured in P-values represents the probability of experiment yielding more extreme results than random chance model.

Note that while p-values are usually compared to significance level of 5%, scientific conclusions and business or policy decisions should not be based only on whether a p-value passes a specified threshold.

For data scientists, p-values are used as intermediate inputs for feature selection of machine learning models.

## Power and Sample Size

Power is the probability of detecting specified effect size with specified sample characteristics (size and variability). In simple terms, power represents probability of correctly rejecting null hypothesis when alternative hypothesis is actually valid.

For computing power or sample size, four main components are required:

1. Sample size

2. Effect size (Minimum desired size of effect to detect in statistical test)

3. Significance level

4. Power

Note that different distribution classes have different power test available. Refer to the documentation below for more details:
https://www.statsmodels.org/dev/stats.html#power-and-sample-size-calculations

```python
import statsmodels.api as sm
import statsmodels.stats.power as sp
# parameters for power analysis of 2 sample T distribution
current_effect = 0.12
effect = sm.stats.proportion_effectsize(current_effect + 0.05, current_effect)
alpha = 0.05
power = 0.8
nobs = 3000

# perform power analysis to solve for sample size
analysis = sp.TTestIndPower()
result = analysis.solve_power(effect_size=effect, alpha=alpha, power=power, alternative='larger')
print(f'Sample Size to achieve power of {power}: {result}')

# perform power analysis to solve for power
analysis = sp.TTestIndPower()
result = analysis.solve_power(effect_size=effect, alpha=alpha, nobs1=nobs, alternative='larger')
print(f'Power for sample size {nobs}: {result}')

# perform power analysis to solve for effect size
analysis = sp.TTestIndPower()
result = analysis.solve_power(power=power, alpha=alpha, nobs1=nobs, alternative='larger')
print(f'Effect size for sample size {nobs} and power of {power}: {result}')
```

```python
# Plotting power curve for different effect sizes
effect_sizes = np.array([0.025, 0.05, 0.1, 0.3, 0.8])
sample_sizes = np.array(range(5, 10000))
analysis.plot_power(dep_var='nobs', nobs=sample_sizes, effect_size=effect_sizes)
plt.show()
```

```python
# Plotting individual power curve for specified parameters
power_list = np.arange(0.8,1,0.001)
current_effect = 0.12
effect = sm.stats.proportion_effectsize(current_effect + 0.05, current_effect)
alpha = 0.05
size = []
for power in power_list:
    size.append(analysis.solve_power(effect_size=effect, alpha=alpha, power=power, alternative='larger'))
plt.plot(size, list(power_list))
plt.show()
```
