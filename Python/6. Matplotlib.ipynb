## Matplotlib

Matplotlib is one of the fundamental Python packages used for data visualization and exploratory data analysis.

Matplotlib also allows highly-customized plots to be created, which can be integrated with web applications.

This notebook will briefly cover the main use of matplotlib python package, as Seaborn is preferred for data visualization and exploratory data analysis (more details in the next notebook)

```python
# Import matplotlib.pyplot package
import matplotlib.pyplot as plt
import numpy as np
```

### Basic plot types in Matplotlib

For exploratory data analysis and data visualization, the most common types of plots used in Matplotlib are as follows:

1. Histogram
2. Scatter plot
3. Box plot
4. Line plot

```python
# Generating random values for data visualization
x = np.random.normal(size=1000)
y = np.arange(30)
z = np.random.randint(200, size=30)
```

```python
# Histogram plot
plt.hist(x,bins=20)
plt.show()
```

```python
# Boxplot
plt.boxplot(x)
plt.show()
```

```python
# Scatter plot
plt.scatter(y,z)
plt.show()
```

```python
# Line plot
plt.plot(y,z)
plt.show()
```

### Features of Matplotlib API

#### Figures and Subplots

Multiple plots can be displayed under the same figure using subplots.

The following arguments can be used in subplots function:
1. nrows: Number of rows of subplots (1 by default)
2. ncols: Number of columns of subplots (1 by default)
3. sharex: All subplots using same x-axis ticks (False by default)
4. sharey: All subplots using same y-axis ticks (False by default)
5. **fig_kw: Additional keywords to subplots when creating figure

```python
fig, ax = plt.subplots(2,2,figsize=(8,8)) # Initialize subplot with nrows, ncols and figsize
ax[0,0].hist(x,bins=20)
ax[0,1].boxplot(x)
ax[1,0].scatter(y,z)
ax[1,1].plot(y,z)
plt.subplots_adjust(wspace=0.2, hspace=0.2) # Adjust spacings around subplots
plt.show()
```

#### Colors, Linestyles and Markers

For more details about the available customization styles, refer to the following list of documentations below for reference:

Documentation for color palettes in matplotlib: https://matplotlib.org/stable/tutorials/colors/colors.html

Documentation for line styles in matplotlib: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

Documentation for marker types in matplotlib: https://matplotlib.org/stable/api/markers_api.html

```python
# Example of basic matplotlib style customization
fig, ax = plt.subplots(2,2,figsize=(8,8)) # Initialize subplot with nrows, ncols and figsize
ax[0,0].hist(x,bins=20, color = 'r')
ax[0,1].boxplot(x)
ax[1,0].scatter(y,z, color = 'magenta', marker="d")
ax[1,1].plot(y,z, color = (0.2,0.56,0.3), linestyle = ":")
plt.subplots_adjust(wspace=0.2, hspace=0.2) # Adjust spacings around subplots
plt.show()
```

#### Ticks, Labels, Titles and Legends

Matplotlib plots can be customized further on ticks positions, ticks labels, plot titles and legends as shown below:

```python
plt.plot(y,z,label = 'line')
plt.xticks([0,10,20,30], labels=[2009,2010,2011,2012]) # Setting x ticks and its labels
plt.xlabel("x axis") # Setting x axis label
plt.yticks([0,50,100,150,200]) # Setting y ticks
plt.ylabel("y axis") # Setting y axis label
plt.legend() # Showing legend from "label" argument
plt.title("Title 1") # Showing title of plot
plt.show()
```

```python
fig, ax = plt.subplots(2,2,figsize=(8,8)) # Initialize subplot with nrows, ncols and figsize

ax[0,0].hist(x,bins=20, color = 'r')
ax[0,0].set_xticks([-3,-2,-1,0,1,2,3]) # Setting x ticks
ax[0,0].set_xticklabels(labels = [0,1,2,3,4,5,6]) # Setting x tick labels
ax[0,0].set_xlabel("First x") # Setting x axis label
ax[0,0].set_ylabel("First y") # Setting y axis label
ax[0,0].set_title("First plot") # Setting title

ax[0,1].boxplot(x,vert=False)
ax[0,1].set_xlabel("Second x")
ax[0,1].set_ylabel("Second y")
ax[0,1].set_title("Second plot")

ax[1,0].scatter(y,z, color = 'magenta', marker="d")
ax[1,0].set_xlabel("Third x")
ax[1,0].set_ylabel("Third y")
ax[1,0].set_title("Third plot")

ax[1,1].plot(y,z, color = (0.2,0.56,0.3), linestyle = ":", label="Trend")
ax[1,1].set_xlabel("Fourth x")
ax[1,1].set_ylabel("Fourth y")
ax[1,1].set_title("Fourth plot")
ax[1,1].legend() # Showing legend from label argument

plt.subplots_adjust(wspace=0.3, hspace=0.4) # Adjust spacings around subplots
fig.suptitle("Overall figure title") # Setting overall figure title
plt.show()
```

#### Annotations

Annotations can also be added on matplotlib plots as shown below:

```python
plt.plot(y,z)
plt.annotate("Global maximum", xy=(np.argmax(z), max(z)+5), xytext=(np.argmax(z), max(z)+30), 
             arrowprops=dict(facecolor='green', headwidth=3, width=4, headlength=4), 
             horizontalalignment='left', verticalalignment='top')
plt.annotate("Global minimum", xy=(np.argmin(z), min(z)-5), xytext=(np.argmin(z), min(z)-30), 
             arrowprops=dict(facecolor='red', headwidth=3, width=4, headlength=4), 
             horizontalalignment='left', verticalalignment='top')
plt.show()
```

#### Saving matplotlib figures

Matplotlib plot figures can also be saved as shown below:

```python
plt.plot(y,z)
plt.annotate("Global maximum", xy=(np.argmax(z), max(z)+5), xytext=(np.argmax(z), max(z)+30), 
             arrowprops=dict(facecolor='green', headwidth=3, width=4, headlength=4), 
             horizontalalignment='left', verticalalignment='top')
plt.annotate("Global minimum", xy=(np.argmin(z), min(z)-5), xytext=(np.argmin(z), min(z)-30), 
             arrowprops=dict(facecolor='red', headwidth=3, width=4, headlength=4), 
             horizontalalignment='right', verticalalignment='top')
plt.savefig("Data_sources/Figure1.png", facecolor=(1, 1, 1)) # Saving matplotlib figures
plt.show()
```
